```python
%load_ext autoreload
%autoreload 2
import pandas as pd
import numpy as np
import scipy
import os, sys
import zipfile

def project_root(folder, parent):
    while os.path.split(folder)[1] not in ('', parent):
        folder, _ = os.path.split(folder)
    return folder

root_folder = project_root(os.getcwd(), 'welfare_state_analytics')
sys.path.append(root_folder)

import westac.corpus.text_corpus as text_corpus
import westac.corpus.corpus_vectorizer as corpus_vectorizer
import types

# df = pd.read_excel('./data/year+text_window.xlsx')
# df.to_csv('./data/year+text_window.txt', sep='\t')
```

```python

source_filename = os.path.join(root_folder, "data", "year+newspaper+text_2019-10-16.txt")

df = pd.read_csv(source_filename, sep='\t') # [['year', 'txt']]
```

```python
import zipfile

target_name = os.path.join(root_folder, "data", "year+newspaper+text_yearly_document.txt.zip")

documents = df.fillna('').groupby(['year', 'newspaper'])['txt'].apply(lambda x: ' '.join(x)).reset_index()

with zipfile.ZipFile(target_name, "w") as zf:

    for index, document in documents.iterrows():

        store_filename = "{}_{}.txt".format(document['newspaper'], document['year'])

        zf.writestr(store_filename, document['txt'], zipfile.ZIP_DEFLATED)
```

```python
df = pd.read_csv('./co_occurrence_1945-1989.csv', sep='\t')
```

```python
for newspaper in df.newspaper.unique():
    df.where(lambda z: z['newspaper'] == newspaper).groupby(['year']).count()['value'].plot()
```

```python
df.groupby(['year']).nunique()[['w1', 'w2']].plot()
```

```python
df['w1+w2'] = df.apply(lambda x: '{} {}'.format(str(x['w1']), str(x['w2'])), axis=1)
```

```python
df.groupby(['year']).nunique()[['w1', 'w2', 'w1+w2']].plot()
```

```python
df.groupby(['year']).count()['w1'].plot()
```
